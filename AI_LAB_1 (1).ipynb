```python
graph = {
  'GUJARAT' : ['MADHYAPRADESH','MAHARASTRA'],
  'MADHYAPRADESH' : ['GUJARAT', 'CHATTISGARH','MAHARASTRA'],
  'CHATTISGARH' : ['MADHYAPRADESH','MAHARASTRA','TELANGANA','JHARKHAND','ODISHA'],
  'JHARKHAND' : ['CHATTISGARH','ODISHA','WESTBENGAL'],
  'WESTBENGAL' : ['JHARKHAND', 'ODISHA'],
  'MAHARASTRA' : ['GUJARAT','MADHYAPRADESH','CHATTISGARH','TELANGANA','KARNATAKA','GOA'],
  'TELANGANA' : ['MAHARASTRA','KARNATAKA','ANDHRAPRADESH','CHATTISGARH'],
  'ANDHRAPRADESH' : ['ODISHA','TELANGANA','TAMILNADU','KARNATAKA'],
  'ODISHA' : ['CHATTISGARH','JHARKHAND','WESTBENGAL','ANDHRAPRADESH'],
  'GOA' : ['KARNATAKA','MAHARASTRA'],
  'KARNATAKA' : ['GOA','MAHARASTRA','TELANGANA','ANDHRAPRADESH','TAMILNADU','KERALA'],
  'KERALA' : ['TAMILNADU', 'KARNATAKA'],
  'TAMILNADU' : ['KARNATAKA','ANDHRAPRADESH','KERALA']
}
```

# **BREATH FIRST SEARCH TRAVERSAL IMPLEMENTATION**

```python
visited = [] # List for visited nodes.
queue = []     #Initialize a queue

def bfs(visited, graph, node): #function for BFS
  visited.append(node)
  queue.append(node)

  while queue:          # Creating loop to visit each node
    m = queue.pop(0) 
    print (m, end = "-->") 

    for neighbour in graph[m]:
      if neighbour not in visited:
        visited.append(neighbour)
        queue.append(neighbour)

# Driver Code
print("BREATH FIRST-SEARCH TRAVERSAL RESULT")
bfs(visited, graph, 'ANDHRAPRADESH')    # function calling
```

# **DEPTH FIRST SERACH TRAVERSAL IMPLEMENTATION**

```python
visited = set() # Set to keep track of visited nodes of graph.

def dfs(visited, graph, node):  #function for DFS
    if node not in visited:
        print (node, end="-->")
        visited.add(node)
        for neighbour in graph[node]:
            dfs(visited, graph, neighbour)

# Driver Code
print("DEPTH FIRST-SEARCH TRAVERSAL RESULT")
dfs(visited, graph, 'MADHYAPRADESH')
```
